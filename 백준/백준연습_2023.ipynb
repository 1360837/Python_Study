```python
#14719 빗물 (1) - 오답 ( Why? )

H, W=map(int,input().split())

B=list(map(int,input().split()))
V=0
k=0

while True:
  for i in range (k+1,W):
    if B[k]<=B[i] and i-k>=2:
      if k==0:
        for j in range(k+1,i):
          if min(B[k],B[i])>B[j]:
            V+=min(B[k],B[i])-B[j]
      else :
        for j in range(k,i):
          if min(B[k-1],B[i])>B[j]:
            V+=min(B[k-1],B[i])-B[j]
      k=i
  k+=1
  if k>=W-1:break

print(V)
```

```python
#14719 빗물 (2)

H, W = map(int,input().split())
Blocks = list(map(int,input().split()))
Rain=0

for i in range(1,W-1):
  LM = max(Blocks[:i])
  RM = max(Blocks[i+1:])

  TM = min(LM,RM)

  if Blocks[i] < TM:
    Rain += TM - Blocks[i]

print(Rain)
```

```python
#2525 오븐 시계

A,B=map(int,input().split())
C=int(input())

if B+C>=60:
  A+=(B+C)//60
  B=(B+C)%60
else:
  B+=C

if A>=24:
  A=A%24

print(A,B)
```

```python
#1110 더하기 사이클

N=int(input())
cnt=0

new=N

while True:
  temp=(new//10)+(new%10)
  new=(new%10)*10+(temp%10)
  cnt+=1
  if N==new:
    break

print(cnt)
```

```python
#1110 더하기 사이클 - 함수 이용

N=int(input())
cnt=0

new=N

def new_num(num):
  temp=(num//10)+(num%10)
  num=(num%10)*10+(temp%10)
  return num

while True:
  new=new_num(new)
  cnt+=1
  if N==new:
    break

print(cnt)
```

```python
#1940 주몽

N=int(input())
M=int(input())
Nums=sorted(list(map(int,input().split())))
L,R=0,N-1
Cnt=0

while L<R:
  Result = Nums[L]+Nums[R]
  if Result<M:
    L+=1
  elif Result>M:
    R-=1
  else:
    Cnt+=1
    L+=1
    R-=1

print(Cnt)
```

```python
#1302 베스트셀러

BN=sorted([input() for _ in range(int(input()))])
cnt=0
index=0

for i in range(len(BN)):
  cnt1=1
  for j in range(i+1,len(BN)-1):
    if BN[i]==BN[j] :
      cnt1+=1
      i=j
  if cnt1>cnt:
    cnt=cnt1
    index=i

print(BN[index])
```

```python
#1158 요세푸스 문제

N,K=map(int,input().split())
p=[i for i in range(1,N+1)]

result=[]
cnt=0

for _ in range(N):
  cnt+=K-1
  if cnt>=len(p):
    cnt%=len(p)
  result.append(str(p.pop(cnt)))

print("<",", ".join(result),">",sep="")
```

```python
#1668 트로피 진열

N=int(input())
h=[]

for i in range(N):
  h.append(int(input()))

leftmax=rightmax=0
leftcnt=rightcnt=0

for i in h:
  if i>leftmax:
    leftcnt+=1
    leftmax=i
for i in h[::-1]:
  if i>rightmax:
    rightcnt+=1
    rightmax=i

print(leftcnt)
print(rightcnt)
```

```python
#2511 카드놀이
ACards=list(map(int,input().split()))
BCards=list(map(int,input().split()))
Ascore=Bscore=0
if ACards==BCards:
  print("10 10")
  print("D")
else:
  for i in range(10):
    if ACards[i]>BCards[i]:
      Ascore+=3
      win="A"
    elif BCards[i]>ACards[i]:
      Bscore+=3
      win="B"
    else:
      Ascore+=1
      Bscore+=1
  print(Ascore,Bscore,sep=" ")
  if Ascore>Bscore:
    print("A")
  elif Bscore>Ascore:
    print("B")
  else :
    print(win)
```

```python
#1769 3의 배수 - 시간초과

X=int(input())
cnt=0

while True:
  Y=0
  while X>0:
    Y+=X%10
    X=X//10
  X=Y
  cnt+=1
  if Y<10:break


print(cnt)
if Y%3==0:
  print("YES")
else:
  print("NO")
```
